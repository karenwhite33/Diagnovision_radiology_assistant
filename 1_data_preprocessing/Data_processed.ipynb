## 0. Pequeño Análisis

```python
# prompt: Cargar los archivos cdv subidos al notebook

import pandas as pd

# Reemplaza 'nombre_del_archivo.csv' con el nombre real de tu archivo
train = pd.read_csv('train_filtered.csv')
val = pd.read_csv('val_filtered.csv')

# Muestra las primeras filas del DataFrame para verificar que se cargó correctamente
print(train.head())
```

## 1. Eliminar filas

```python

# Eliminar filas donde 'frontal_lateral' sea 'lateral' en el DataFrame 'train'
train = train[train['frontal_lateral'] != 'lateral']

# Eliminar filas donde 'frontal_lateral' sea 'lateral' en el DataFrame 'val'
val = val[val['frontal_lateral'] != 'lateral']
```

```python
# Eliminar filas donde 'ap_pa' sea 'pa' en el DataFrame 'train'
train = train[train['ap_pa'] != 'pa']

# Eliminar filas donde 'ap_pa' sea 'pa' en el DataFrame 'val'
val = val[val['ap_pa'] != 'pa']
```

```python
# Eliminar duplicados en la columna 'deid_patient_id' manteniendo la primera aparición
train = train.drop_duplicates(subset=['deid_patient_id'], keep='first')
val = val.drop_duplicates(subset=['deid_patient_id'], keep='first')

# Verificar el resultado (opcional)
print(train['deid_patient_id'].nunique())
print(val['deid_patient_id'].nunique())
```

## 2. Eliminar columnas innecesarias

```python
# prompt: dame los nombres de las columnas del df

print(train.columns)
val.columns
```

```python
# prompt: necesito eliminar varias columnas del dataset

import pandas as pd

# Lista de columnas a eliminar
columnas_a_eliminar = ["path_to_image", "frontal_lateral", "ap_pa", "patient_report_date_order", "report", "section_narrative", "section_clinical_history", "section_history", "section_comparison", "section_technique", "section_procedure_comments", "section_findings", "section_end_of_impression", "section_summary", "section_accession_number", "age", "sex", "race", "ethnicity", "interpreter_needed", "insurance_type", "recent_bmi", "deceased", "split", "No Finding"]

# Elimina las columnas del DataFrame train
train = train.drop(columns=columnas_a_eliminar, errors='ignore')

# Elimina las columnas del DataFrame val
val = val.drop(columns=columnas_a_eliminar, errors='ignore')

# Muestra las primeras filas del DataFrame para verificar que se cargó correctamente
print(train.head())
print(val.head())

train.info()
train.head(5)
```

```python
print(train.columns)
```

## 3. Modificar valores de patologías

```python
# Columns where you want to replace NaN with 0
columns_to_fill = ["Enlarged Cardiomediastinum", "Cardiomegaly", "Lung Opacity", "Lung Lesion", "Edema", "Consolidation", "Pneumonia", "Atelectasis", "Pneumothorax", "Pleural Effusion", "Pleural Other", "Fracture"] # Replace with your actual column names

# Fill NaN values in specified columns with 0
train[columns_to_fill] = train[columns_to_fill].fillna(0)
val[columns_to_fill] = val[columns_to_fill].fillna(0)
```

```python
train.head(5)
```

## 4. Unir DataFrames y split aleatorio

```python
import pandas as pd
# Concatenate train and val DataFrames
df_combined = pd.concat([train, val], ignore_index=True)

# Now df_combined contains all the data from both train and val
print(df_combined.head())
print(df_combined.info())
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
train_df, val_df = train_test_split(df_combined, test_size=0.1, random_state=42)

print(train_df.head())
print(val_df.head())

print(train_df.shape)
val_df.shape
```

```python
train_df.info()
val_df.info()
```

```python

from google.colab import files

train_df.to_csv('train_processed.csv', index=False)
val_df.to_csv('val_processed.csv', index=False)

files.download('train_processed.csv')
files.download('val_processed.csv')
```
